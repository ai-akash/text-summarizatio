```python
import pandas as pd
import re
import string
from string import punctuation
```

```python
data =pd.read_csv("data/data_without_outlier.csv",engine = "python",encoding = "utf-8")
display(data.head())
display(data.info())
```

```python
punctuation+= "।"
```

```python
data["summary"] = data["summary"].map(lambda s :s.translate(str.maketrans('', '', punctuation)))
data["article"] = data["article"].map(lambda s :s.translate(str.maketrans('', '', punctuation)))
```

```python
data["summary"] = data["summary"].map(lambda s : " ".join(s.split()))
data["article"] = data["article"].map(lambda s : " ".join(s.split()))
```
